###Following section contain the various tests conducted on the data



```python
#Chi Square Test Of Independence
#H0 = There is no correlation between age group and payment type preferred
df1 = pd.crosstab(df['Practical_Experience'], df['Job_Readiness_Confidence'])
print(df1)

stat, p, dof, expected = stats.chi2_contingency(df1)
crit = stats.chi2.ppf(q = 0.95, # Find the critical value for 95% confidence*
                      df = 6)

print("\nCritical value: ", crit)
print("Test statistic: ", stat)
print("P-value: ", p)
print("There is no significant association between job readiness confidence and practical experience")
#Note: The degrees of freedom for a test of independence equals the product of the number of categories in each variable minus 1.
```

```python
#Chi Square Test of Independence
#H0 = There is no correlation between education satisfaction and teaching method effectiveness
df1 = pd.crosstab(df['Education_Satisfaction'], df['TM_Effectiveness'])
print(df1)

stat, p, dof, expected = stats.chi2_contingency(df1)
crit = stats.chi2.ppf(q = 0.95, # Find the critical value for 95% confidence*
                      df = 6)

print("\nCritical value: ", crit)
print("Test statistic: ", stat)
print("P-value: ", p)
print("There is significant association between education satisfaction and effectiveness of teaching methods")
#Note: The degrees of freedom for a test of independence equals the product of the number of categories in each variable minus 1.
```

```python
#Anova Test to determine whether students enrolled in different programs have similar education satisfaction opinion or not
a = df.loc[df['Field of Study'] == "MSC BDA",'Education_Satisfaction'].to_list()
b = df.loc[df['Field of Study'] == "IT","Education_Satisfaction"].to_list()
c = df.loc[df['Field of Study'] == "Pure Sciences", 'Education_Satisfaction'].to_list()
d = df.loc[df['Field of Study'] == "Business Oriented (BBI, BAF, BFM, BMM, BMS and related)",'Education_Satisfaction'].to_list()
e = df.loc[df['Field of Study'] == "Communication, Journalism, Entertainment, Media Advertising and related",'Education_Satisfaction'].to_list()
f = df.loc[df['Field of Study'] == "Medical",'Education_Satisfaction'].to_list()
g = df.loc[df['Field of Study'] == "Architecture ",'Education_Satisfaction'].to_list()
h = df.loc[df['Field of Study'] == "Engineering",'Education_Satisfaction'].to_list()
i = df.loc[df['Field of Study'] == "Law",'Education_Satisfaction'].to_list()

res = stats.f_oneway(a,b,c,d,e,f,g,h,i)
print("Result: ", res)

print("Since the pvalue is less than 0.05, H1 accepted")
print("It is evident that student in different programs have different outlooks towards education satisfaction")
```

```python
#Anova Test to determine whether education statisfaction of students vary by location or not
a = df.loc[df['CollegeLocation'] == "Mumbai City",'Education_Satisfaction'].to_list()
b = df.loc[df['CollegeLocation'] == "Mumbai Suburban","Education_Satisfaction"].to_list()
c = df.loc[df['CollegeLocation'] == "Navi Mumbai", 'Education_Satisfaction'].to_list()
d = df.loc[df['CollegeLocation'] == "Thane",'Education_Satisfaction'].to_list()

res = stats.f_oneway(a,b,c,d)
print("Result: ", res)

print("Since the pvalue is greater than 0.05, H0 accepted")
print("It is evident that education satisfaction of students among colleges across difference locations do not have significant variation")
```

```python
#Ftest for variance of job readiness between groups that have practical experience and those who don't
a = df.loc[df['Practical_Experience'] == "Yes",'Job_Readiness_Confidence'].to_list()
b = df.loc[df['Practical_Experience'] == "No",'Job_Readiness_Confidence'].to_list()

x = np.array(a)
y = np.array(b)

print(np.var(a), np.var(b))

def f_test(a, b):
   f = np.var(a, ddof=1)/np.var(b, ddof=1)
   nun = x.size-1
   dun = y.size-1
   p_value = 1-stats.f.cdf(f, nun, dun)
   return f, p_value

# perform F-test
f,p = f_test(x, y)
print("F-statistic: ",f)
print("P-value: ",p)
print("Since the p-value is >0.05, hence the null hypothesis is accepted.\nThere is no significant differences in the variance of the groups")
```
